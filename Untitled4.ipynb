### Load Main Dataset for Training and Validation

Now, we'll load the full IAM dataset with synthesized cross-outs from the unzipped directory. This will be used for the primary training and validation of the deep learning model. The `base_dataset_path` is set to the extracted location.

*Note: The previous code block above this one, which loaded the `/content/drive/MyDrive/img/` dataset, will be repurposed later for the 'small-scale experimentation' evaluation.*

```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt

# Define the base path for the full IAM dataset (already unzipped)
base_dataset_path = 'E:\project\dataset'

if not os.path.exists(base_dataset_path):
    print(f"Error: Base dataset path '{base_dataset_path}' not found. Please ensure the zip file was extracted correctly.")
else:
    print(f"Using base dataset path for main datasets: '{base_dataset_path}'")

# Use existing image_height, image_width, batch_size from previous setup
image_height = 100 # Assuming these are defined elsewhere or will be defined globally
image_width = 200
batch_size = 8

def load_dataset_split_main(split_name, label_mode='categorical'):
    split_path = os.path.join(base_dataset_path, split_name, 'images')
    if not os.path.exists(split_path):
        print(f"Warning: Split path '{split_path}' not found. Skipping {split_name} dataset loading.")
        return None

    print(f"Loading {split_name} dataset from: {split_path}")
    dataset = tf.keras.utils.image_dataset_from_directory(
        split_path,
        labels='inferred',
        label_mode=label_mode,
        image_size=(image_height, image_width),
        interpolation='lanczos3',
        batch_size=batch_size,
        shuffle=(split_name == 'train'), # Only shuffle training data
        seed=42 if split_name == 'train' else None # Set seed for reproducibility during training shuffle
    )
    print(f"  {split_name} dataset created with {len(dataset.class_names)} classes: {dataset.class_names}")
    print(f"  Total batches: {len(dataset)}")
    return dataset

# Load the training, validation, and test datasets separately for the main model
train_ds_full = load_dataset_split_main('train')
val_ds_full = load_dataset_split_main('val') # Corrected: 'validation' to 'val'
test_ds_full = load_dataset_split_main('test')

# Keep track of class names from the training dataset for consistency
if train_ds_full:
    class_names = train_ds_full.class_names # This will be the definitive list of class names
    num_classes = len(class_names)
    print(f"Detected {num_classes} classes for main dataset: {class_names}")
else:
    print("Could not load main training dataset. Please check the base_dataset_path and extraction.")
    class_names = []
    num_classes = 0

# Example visualization from the full training dataset
if train_ds_full:
    plt.figure(figsize=(10, 10))
    plt.suptitle("Sample Images from Main Training Dataset", fontsize=16)
    for images, labels in train_ds_full.take(1):
        for i in range(min(9, len(images.numpy()))):
            ax = plt.subplot(3, 3, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            predicted_class_index = np.argmax(labels[i].numpy())
            plt.title(class_names[predicted_class_index])
            plt.axis("off")
    plt.show()
```

```python
dataset_size = len(dataset)
train_size = int(0.7 * dataset_size)
val_size = int(0.15 * dataset_size)
test_size = dataset_size - train_size - val_size

train_ds = dataset.take(train_size)
remaining_ds = dataset.skip(train_size)
val_ds = remaining_ds.take(val_size)
test_ds = remaining_ds.skip(val_size)

print(f"Dataset split into: ")
print(f"  Training: {len(train_ds)} batches")
print(f"  Validation: {len(val_ds)} batches")
print(f"  Test: {len(test_ds)} batches")
```

```python
AUTOTUNE = tf.data.AUTOTUNE

# Prepare the full datasets with caching and prefetching
train_ds_full = train_ds_full.shuffle(100).prefetch(buffer_size=AUTOTUNE)
val_ds_full = val_ds_full.prefetch(buffer_size=AUTOTUNE)
test_ds_full = test_ds_full.prefetch(buffer_size=AUTOTUNE)
    
print("Full datasets prepared with caching and prefetching.")
```

```python
import tensorflow as tf

tf.keras.mixed_precision.set_global_policy('mixed_float16')
num_classes = len(class_names) # Correctly use class_names from the full dataset
print(f"Number of classes for model output: {num_classes}")

model = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1./255, input_shape=(image_height, image_width, 3)),
  tf.keras.layers.Conv2D(16, 3, padding='same', activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, padding='same', activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(64, 3, padding='same', activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

model.summary()
```

```python
epochs = 10
history = model.fit(
    train_ds_full,
    validation_data=val_ds_full,
    epochs=epochs
)

print("Model training complete with full dataset.")
```

```python
print("Evaluating model on the test dataset...")
loss, accuracy = model.evaluate(test_ds_full)

print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
