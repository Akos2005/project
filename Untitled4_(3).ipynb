```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt

base_dataset_path = 'E:\\project\\dataset'

image_height = 100
image_width = 200
batch_size = 128

def load_dataset_split_main(split_name, label_mode='categorical'):
    split_path = os.path.join(base_dataset_path, split_name, 'images')
    if not os.path.exists(split_path):
        print(f"Warning: Split path '{split_path}' not found. Skipping {split_name} dataset loading.")
        return None

    print(f"Loading {split_name} dataset from: {split_path}")
    dataset = tf.keras.utils.image_dataset_from_directory(
        split_path,
        labels='inferred',
        label_mode=label_mode,
        image_size=(image_height, image_width),
        interpolation='lanczos3',
        batch_size=batch_size,
        shuffle=(split_name == 'train'),
        seed=42 if split_name == 'train' else None
    )
    print(f"  {split_name} dataset created with {len(dataset.class_names)} classes: {dataset.class_names}")
    print(f"  Total batches: {len(dataset)}")
    return dataset

train_ds_full = load_dataset_split_main('train')
val_ds_full = load_dataset_split_main('val')
test_ds_full = load_dataset_split_main('test')


if train_ds_full:
    class_names = train_ds_full.class_names
    num_classes = len(class_names)
else:
    class_names = []
    num_classes = 0

if train_ds_full:
    plt.figure(figsize=(12, 12))
    plt.suptitle("Sample Images")

    selected_images = []
    selected_labels = []
    found_classes = set()

    for images, labels in train_ds_full.take(num_classes * 2):
        for i in range(len(images)):
            predicted_class_index = np.argmax(labels[i].numpy())
            class_name = class_names[predicted_class_index]

            if class_name not in found_classes:
                selected_images.append(images[i].numpy().astype("uint8"))
                selected_labels.append(class_name)
                found_classes.add(class_name)

            if len(found_classes) == num_classes:
                break
        if len(found_classes) == num_classes:
            break


    for i in range(min(num_classes, len(selected_images))):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(selected_images[i])
        plt.title(selected_labels[i])
        plt.axis("off")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
dataset_size = len(dataset)
train_size = int(0.7 * dataset_size)
val_size = int(0.15 * dataset_size)
test_size = dataset_size - train_size - val_size

train_ds = dataset.take(train_size)
remaining_ds = dataset.skip(train_size)
val_ds = remaining_ds.take(val_size)
test_ds = remaining_ds.skip(val_size)

print(f"Dataset split into: ")
print(f"  Training: {len(train_ds)} batches")
print(f"  Validation: {len(val_ds)} batches")
print(f"  Test: {len(test_ds)} batches")
```

```python
AUTOTUNE = tf.data.AUTOTUNE

# Prepare the full datasets with caching and prefetching
train_ds_full = train_ds_full.shuffle(100).prefetch(buffer_size=AUTOTUNE)
val_ds_full = val_ds_full.prefetch(buffer_size=AUTOTUNE)
test_ds_full = test_ds_full.prefetch(buffer_size=AUTOTUNE)

print("Full datasets prepared with caching and prefetching.")
```

```python
import tensorflow as tf

tf.keras.mixed_precision.set_global_policy('mixed_float16')
num_classes = len(class_names) # Correctly use class_names from the full dataset
print(f"Number of classes for model output: {num_classes}")

model = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1./255, input_shape=(image_height, image_width, 3)),
  tf.keras.layers.Conv2D(16, 3, padding='same', activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Conv2D(32, 3, padding='same', activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Conv2D(64, 3, padding='same', activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['accuracy'])

model.summary()
```

```python
import tensorflow as tf

print("Verzió:", tf.__version__)
print("Elérhető eszközök:", tf.config.list_physical_devices())
gpu_devices = tf.config.list_physical_devices('GPU')
if gpu_devices:
    print(f"GPU elérhető: {gpu_devices}")
else:
    print("A TensorFlow NEM látja a GPU-t. CPU-n fog futni.")
```

```python
epochs = 40
history = model.fit(
    train_ds_full,
    validation_data=val_ds_full,
    epochs=epochs
)

print("Model training complete with full dataset.")
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='training accuracy')
plt.plot(epochs_range, val_acc, label='validation accuracy')
plt.legend(loc='lower right')
plt.title('training and validation accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='training loss')
plt.plot(epochs_range, val_loss, label='validation loss')
plt.legend(loc='upper right')
plt.title('training and validation loss')
plt.show()
```

```python
print("Evaluating model on the test dataset...")
loss, accuracy = model.evaluate(test_ds_full)

print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

for images, labels in test_ds_full.take(1):
    predictions = model.predict(images)

    plt.figure(figsize=(15, 15))
    plt.suptitle("Model Predictions on Test Images", fontsize=16)

    for i in range(min(9, len(images))):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))

        true_label_index = np.argmax(labels[i].numpy())
        predicted_label_index = np.argmax(predictions[i])
        confidence = np.max(predictions[i]) * 100

        true_label_name = class_names[true_label_index]
        predicted_label_name = class_names[predicted_label_index]

        color = "green" if true_label_index == predicted_label_index else "red"

        plt.title(f"True: {true_label_name}\nPred: {predicted_label_name} ({confidence:.1f}%) ", color=color)
        plt.axis("off")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
```

```python
import sklearn.metrics
import seaborn as sns

y_true = []
y_pred = []

for images, labels in test_ds_full:
    predictions = model.predict(images)
    y_true.extend(np.argmax(labels.numpy(), axis=1))
    y_pred.extend(np.argmax(predictions, axis=1))

cm = sklearn.metrics.confusion_matrix(y_true, y_pred)
    
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
